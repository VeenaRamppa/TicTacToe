```python
print(" Welcome to Tic Tac Toe game......")
import pdb
player1_symbol = ""
player2_symbol = ""
lock_moves = [0]
board_marking = ["#","1","2","3","4","5","6","7","8","9"]
set_flag = False

def player_symbol():
    global player1_symbol
    global player2_symbol
    player1_symbol = input("Player 1 which symbol u like to use...X or O ? ")
    if player1_symbol == "X":
         player2_symbol = "O"
        
    else:
        player2_symbol = "X"
        

def print_in_firstrow(move,player):
    global player1_symbol
    global player2_symbol
    global board_marking
    
    if move == 1:
            if player == 1:                
                board_marking[1] = player1_symbol
            else:
                board_marking[1] = player2_symbol
                
            print(f"| {board_marking[1]} | {board_marking[2]} | {board_marking[3]} |")
            
    elif move == 2:
            if player == 1:
                board_marking[2] = player1_symbol
            else:
                board_marking[2] = player2_symbol            
            
            print(f"| {board_marking[1]} | {board_marking[2]} | {board_marking[3]} |")
            
    elif move == 3:
            if player == 1:
                board_marking[3] = player1_symbol
            else:
                board_marking[3] = player2_symbol
            
            print(f"| {board_marking[1]} | {board_marking[2]} | {board_marking[3]} |")

def print_in_secondrow(move,player):
    global player1_symbol
    global player2_symbol
    global board_marking
    
    if move == 4:
            if player == 1:
                board_marking[4] = player1_symbol
            else:
                board_marking[4] = player2_symbol
                
            print(f"| {board_marking[4]} | {board_marking[5]} | {board_marking[6]} |")
            
    elif move == 5:
            if player == 1:
                board_marking[5] = player1_symbol
            else:
                board_marking[5] = player2_symbol            
            
            print(f"| {board_marking[4]} | {board_marking[5]} | {board_marking[6]} |")
            
    elif move == 6:
            if player == 1:
                board_marking[6] = player1_symbol
            else:
                board_marking[6] = player2_symbol
            
            print(f"| {board_marking[4]} | {board_marking[5]} | {board_marking[6]} |")

def print_in_thirdrow(move,player):
    global player1_symbol
    global player2_symbol
    global board_marking
    
    if move == 7:
            if player == 1:
                board_marking[7] = player1_symbol
            else:
                board_marking[7] = player2_symbol
                
            print(f"| {board_marking[7]} | {board_marking[8]} | {board_marking[9]} |")
            
    elif move == 8:
            if player == 1:
                board_marking[8] = player1_symbol
            else:
                board_marking[8] = player2_symbol            
            
            print(f"| {board_marking[7]} | {board_marking[8]} | {board_marking[9]} |")
            
    elif move == 9:
            if player == 1:
                board_marking[9] = player1_symbol
            else:
                board_marking[9] = player2_symbol
            
            print(f"| {board_marking[7]} | {board_marking[8]} | {board_marking[9]} |")

def chk_for_match(move,player):
    
    global board_marking
    global set_flag

    if (board_marking[1] == board_marking[2]) and ( board_marking[2] == board_marking[3]):
            print(f"Player {player} won ... ")
            set_flag = True        
    
    elif(board_marking[1] == board_marking[4]) and ( board_marking[4] == board_marking[7]):
            print(f"Player {player} won ...")
            set_flag = True    
    elif(board_marking[1] == board_marking[5]) and ( board_marking[5] == board_marking[9]):
            print(f"Player {player} won ...")
            set_flag = True
    elif(board_marking[2] == board_marking[5]) and (board_marking[5] == board_marking[8]):            
            print(f"Player {player} won ...")
            set_flag = True
    elif(board_marking[3] == board_marking[6]) and ( board_marking[6] == board_marking[9]):
            print(f"Player {player} won ...")
            set_flag = True
    elif(board_marking[3] == board_marking[5]) and ( board_marking[5] == board_marking[7]):
            print(f"Player {player} won ...")
            set_flag = True
    elif(board_marking[4] == board_marking[5]) and (board_marking[5] == board_marking[6]):    
            print(f"Player {player} won ...")
            set_flag = True
    elif(board_marking[7] == board_marking[8]) and ( board_marking[8] == board_marking[9]):    
            print(f"Player {player} won ...")
            set_flag = True
    else:
        print("Try with other move.....")
        
        

def display_board(move=0,player = 0):
    global board_marking
    
    if move == 0 and player == 0:
        print("---------")
        print(f"| {board_marking[1]} | {board_marking[2]} | {board_marking[3]} |")
        print("---------")
        print(f"| {board_marking[4]} | {board_marking[5]} | {board_marking[6]} |")
        print("---------")
        print(f"| {board_marking[7]} | {board_marking[8]} | {board_marking[9]} |")
        print("---------")
    elif (move == 1) or (move == 2) or (move == 3):
        
        print("---------------")
        print_in_firstrow(move,player)
        print("---------------")
        print(f"| {board_marking[4]} | {board_marking[5]} | {board_marking[6]} |")
        print("---------------")
        print(f"| {board_marking[7]} | {board_marking[8]} | {board_marking[9]} |")
        print("---------------")
        
    elif (move == 4) or (move == 5) or (move == 6):
        print("---------------")
        print(f"| {board_marking[1]} | {board_marking[2]} | {board_marking[3]} |")
        print("---------------")
        print_in_secondrow(move,player)
        print("---------------")
        print(f"| {board_marking[7]} | {board_marking[8]} | {board_marking[9]} |")
        print("---------------")
        
    elif (move == 7) or (move == 8) or (move == 9):
        print("---------------")
        print(f"| {board_marking[1]} | {board_marking[2]} | {board_marking[3]} |")
        print("---------------")
        print(f"| {board_marking[4]} | {board_marking[5]} | {board_marking[6]} |")
        print("---------------")
        print_in_thirdrow(move,player)
        print("---------------")
        
    if player != 0:
        chk_for_match(move,player)   
    
player_symbol()

print("----------- Let's Begin.... -----------")

display_board()
while True:
    #pdb.set_trace()
    p_move = int(input("Player 1 enter your move :"))
            
    if lock_moves.count(p_move) > 0: # ## lock_moves list to hold postions entered by the player . it is used to avoid duplicate entry to the same position
        print("Position is locked ... try someother position")
        p_move = int(input("Player 1 enter your move :"))
        continue
    else:
        lock_moves.append(p_move)
        display_board(p_move,1)        
        if set_flag == True:
            break

    p_move = int(input("Player 2 enter your move :"))
    if lock_moves.count(p_move) > 0:
        print("Position is locked ... try someother position")
        p_move = int(input("Player 2 enter your move :"))
        continue
    else:
        lock_moves.append(p_move)
        display_board(p_move,2)        
        if set_flag == True:
            break
```
